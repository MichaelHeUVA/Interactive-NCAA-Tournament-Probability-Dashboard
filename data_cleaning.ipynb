```python
import pandas as pd
```

```python
# load the data from the CSV file datatypes are defined for the columns that had errors
data = pd.read_csv(
    "./data/match_data.csv",
    dtype={
        89: str,
        97: str,
        98: str,
    },
    parse_dates=[91],
)
data.head()
```

```python
# show the info and description of the data
data.info()
data.describe()
```

```python
# drop all the columns that are not useful
data = data.drop("Club ID", axis=1)
data = data.drop("Club Name", axis=1)
data = data.drop("Competition Name", axis=1)
data = data.drop("Period Number", axis=1)
data = data.drop("Period Type", axis=1)
data = data.drop("Assists Defensive", axis=1)
data = data.drop("Biggest Lead Score", axis=1)
data = data.drop("Biggest Scoring Run Score", axis=1)
data = data.drop("Efficiency Custom", axis=1)
data = data.drop("Fouls Coach Disqualifying", axis=1)
data = data.drop("Fouls Technical Other", axis=1)
data = data.drop("Fouls Throw In", axis=1)
data = data.drop("Fouls Unsportsmanlike", axis=1)
data = data.drop("Head Coach Challenge Accepted", axis=1)
data = data.drop("Head Coach Challenge Rejected", axis=1)
data = data.drop("Basketball Matches - Match → Status", axis=1)
data = data.drop("Basketball Matches - Match → Extra Periods Used", axis=1)
data = data.drop("Basketball Matches - Match → Match Time Utc", axis=1)
data = data.drop("Basketball Matches - Match → Attendance", axis=1)
data = data.drop("Basketball Matches - Match → Weather", axis=1)
data = data.drop("Basketball Matches - Match → Is Conference", axis=1)
data = data.drop("Basketball Matches - Match → Tournament ID", axis=1)
data = data.drop("Basketball Matches - Match → Full Time Played", axis=1)
data = data.drop("Basketball Matches - Match → Home Name", axis=1)
data = data.drop("Basketball Matches - Match → Away Name", axis=1)
data = data.drop("Basketball Matches - Match → Home ID", axis=1)
data = data.drop("Basketball Matches - Match → Away ID", axis=1)
data = data.drop("Basketball Matches - Match → Home Score", axis=1)
data = data.drop("Basketball Matches - Match → Away Score", axis=1)
data = data.drop("Basketball Matches - Match → Include In Conf Stats", axis=1)
data = data.drop("Basketball Matches - Match → Include In Standings", axis=1)
data = data.drop("Basketball Matches - Match → Place If Won", axis=1)
data = data.drop("Basketball Matches - Match → Place If Lost", axis=1)
data.drop_duplicates(inplace=True)
```

```python
# group the data by match id and make sure there are only 2 teams per match
data.sort_values(by=["Match ID", "Points"], inplace=True, ascending=[True, False])
match_counts = data["Match ID"].value_counts()
valid_match_ids = match_counts[match_counts == 2].index
data = data[data["Match ID"].isin(valid_match_ids)]
```

```python
# create a new column that indicates if the team won the match
data["Win"] = data["Points"] > data["Points Against"]
data["Win"] = data["Win"].astype(int)
```

```python
# show the info for the data
data.info()
```

```python
# save the cleaned data
data.to_csv("data/match_data_cleaned.csv", index=False)
```
